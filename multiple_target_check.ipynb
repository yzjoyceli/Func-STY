```python
import pandas as pd
import numpy as np
from pandas import DataFrame
import Bio
from Bio import SeqIO
from Bio.Seq import Seq
from Bio.SeqRecord import SeqRecord
from Bio.Alphabet import IUPAC
```

```python
sg = pd.read_csv('../sg_STY/sg_tyr.txt',sep='\t')
```

```python
sg.head()
```

```python
sg.loc[:,'sg_pam'] = sg['sg_seq'].str.cat(sg['pam'])
```

```python
sg.head()
```

```python
sg['sg_pam'].unique().shape
```

```python
out_fa = open("../data/tyr/y_sg_pam.fasta", "w")
```

```python
for index,data in sg.iterrows():
    record = SeqRecord(Seq(data['sg_pam'],IUPAC.unambiguous_dna),id=data['sg_id'])
    SeqIO.write(record,out_fa,'fasta')
```
